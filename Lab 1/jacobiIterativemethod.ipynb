The **Jacobi iterative method** is considered as an iterative algorithm which is used for determining the solutions for the system of **linear equations** in numerical linear algebra, which is diagonally dominant. In this method, an approximate value is filled in for each diagonal element.

We will take a 3 variable **Linear equation** as an example

$$ 15x + 3y + 4z = 9 $$  
$$ 3x + 12y + 10z = 12 $$  
$$ 7x + 2y + 22z = -3 $$ 



Using a **Calculator** we get the values of the variables as,

$$ x = 0.4625 $$ 
$$ y = 1.2125 $$
$$ z = -0.39375 $$

Now we will use the **Jacobi Iterative Method** to calculate the values of the variables.

### Taking inputs of the 3 variable expressions co efficients


```python
x = [0,0,0]
y = [0,0,0]
z = [0,0,0]
c = [0,0,0]

print("Please enter the co efficients of the expressions: ");
for i in range(3):
    print("For the expression ", i+1);
    print("x: ");
    x[i] = int(input());
    print("y: ");
    y[i] = int(input());
    print("z: ");
    z[i] = int(input());
    print("constants: ");
    c[i] = int(input());

for i in range(3):
    print("The expression ", i+1, " is: ", x[i], "x + ", y[i], "y + ", z[i], "z" , " = ", c[i]);
```

### Now following the iterative method we will get an expression for each variables

Swap function to easily swap variable co-efficients

```python
def swap(a,b):
    temp = a;
    a = b;
    b = temp;
    return a,b;
```

In order to be solvable, the equations need to dependent of each other. So the elements of their main diagonals need to the highest among their rows. We will interchange rows if this property is achievable. If it's not we will exit early and terminate the program as the equations can't be solved.

```python
if(x[0] >= x[1]):
    if(x[0] < x[2]):
        x[0],x[2] = swap(x[0],x[2]);
        y[0],y[2] = swap(y[0],y[2]);
        z[0],z[2] = swap(z[0],z[2]);
        c[0],c[2] = swap(c[0],c[2]);

else:
    if(x[1] >= x[2]):
        x[1],x[0] = swap(x[1],x[0]);
        y[1],y[0] = swap(y[1],y[0]);
        z[1],z[0] = swap(z[1],z[0]);
        c[1],c[0] = swap(c[1],c[0]);
    else:
        x[0],x[2] = swap(x[0],x[2]);
        y[0],y[2] = swap(y[0],y[2]);
        z[0],z[2] = swap(z[0],z[2]);
        c[0],c[2] = swap(c[0],c[2]);
        
if(y[1]< y[2]):
        x[1],x[2] = swap(x[1],x[2]);
        y[1],y[2] = swap(y[1],y[2]);
        z[1],z[2] = swap(z[1],z[2]);
        c[1],c[2] = swap(c[1],c[2]);    
    
for i in range(3):
    print("The expression ", i+1, " is: ", x[i], "x + ", y[i], "y + ", z[i], "z" , " = ", c[i]);

if((x[0] < x[1] or x[0] < x[2]) or (y[1] < y[0] or y[1] < y[2]) or (z[2] < z[0] or z[2] < z[1])):
    print("The given equations are consistent yet independent, so not solvable");
    exit(0);
else:
    print("The given equations are dependable,so they are solvable")
```

If the equations are solvable then we will approach to solve them.

```python
x_solve = [];
dx = [];
i = 0;
x_solve.append(((c[i] - (y[i] * 0) - (z[i] * 0)) / x[i]));
dx.append(abs(x_solve[-1]-0));
print("x = ", x_solve[0]);
```

```python
y_solve = [];
dy = [];
i=1;
y_solve.append(((c[i] - (x[i] * 0) - (z[i] * 0)) / y[i]));
dy.append(abs(y_solve[-1]-0));
print("y = ", y_solve[0]);
```

```python

z_solve = [];
dz = [];
i=2;
z_solve.append(((c[i] - (x[i] * 0) - (y[i] * 0)) / z[i]));
dz.append(abs(z_solve[-1]-0));
print("z = ", z_solve[0]);
```

##### Now these are the primary values of the solves. We need to individually solve each variables multiple times to get a value close to the actual value.

```python
for j in range(19):
    for i in range(3):
        # print("The expression ", i+1, " is: ", x[i], "x + ", y[i], "y + ", z[i], "z" , " = ", c[i]);
        # print("x = ", x_solve[j], "y = ", y_solve[j], "z = ", z_solve[j]);
        if(i==0):
            var = ((c[i] - (y[i] * y_solve[j]) - (z[i] * z_solve[j])) / x[i]);
            x_solve.append(var);
            dx.append(abs(x_solve[j+1]-x_solve[j]));
            print("x = ", x_solve[j+1]);
        elif(i==1):
            var = ((c[i] - (x[i] * x_solve[j]) - (z[i] * z_solve[j])) / y[i]);
            y_solve.append(var);
            dy.append(abs(y_solve[j+1]-y_solve[j]));
            print("y = ", y_solve[j+1]);
        else:
            var = ((c[i] - (x[i] * x_solve[j]) - (y[i] * y_solve[j])) / z[i]);
            z_solve.append(var);
            dz.append(abs(z_solve[j+1]-z_solve[j]));
            print("z = ", z_solve[j+1]);
        
    print("--------------------")
    
```

#### Now we can view the values in a table

```python
import matplotlib.pyplot as plt
import numpy as np

columns = ('iteration','x', 'y', 'z', '|dx|', '|dy|', '|dz|');
x_np = np.around(np.array(x_solve), decimals=6);
y_np = np.around(np.array(y_solve), decimals=6);
z_np = np.around(np.array(z_solve), decimals=6);
dx_np = np.around(np.array(dx), decimals=6); 
dy_np = np.around(np.array(dy), decimals=6);
dz_np = np.around(np.array(dz), decimals=6);
iteration = np.array(np.arange(1,len(x_solve)+1));


table = np.column_stack((iteration,x_np,y_np,z_np,dx_np,dy_np,dz_np));

fig, ax = plt.subplots()
ax.axis('off')
ax.axis('tight')
color1 = "#DDBBFF"  # The color you want to apply to all cells
colcolours = [color1 for _ in range(7)]

tab = ax.table(cellText=table, colLabels=columns, colColours = colcolours, cellLoc = 'center', loc='center')
tab.set_fontsize(20)
tab.scale(4, 4)
plt.show()
```

We can see at the last parts of the iterations, the differences are close to zero, so we can say they have actually converged at such point.

```python
limit = 1e-6;
flag = False
for i in range(len(dx)):
    if(dx[i] < limit or dy[i] < limit and dz[i] < limit):
        print("The solution converges at iteration: ", i+1);
        flag = True;
        break;
if(flag == False):
    print("The solution converges at iteration: 20");
```
